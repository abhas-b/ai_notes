```python
orders_list = [("01", "02", 1, 350),
               ("01", "04", 1, 580),
               ("02", "03", 1, 450)]


orders_df = spark.createDataFrame(orders_list).toDF("order_id", "prod_id", "qty", "unit_price")
```

```python
product_list = [("02", "Scoll Mouse", 250, 20),
                ("04", "Keybord", 100, 15)]

product_df = spark.createDataFrame(product_list).toDF("prod_id", "prod_name", "list_price", "qty")
```

```python
join_expr = orders_df.prod_id == product_df.prod_id
```

```python
join_df = orders_df.join(product_df,
                         join_expr,
                         "inner")

display(join_df)
```

```python
import pyspark.sql.functions as F

join_df = orders_df.join(product_df, join_expr,"outer")\
                             .withColumn("prod_name", F.coalesce(product_df.prod_name, orders_df.prod_id))

display(join_df)

```
