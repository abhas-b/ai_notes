# Caching Basics

```sql
-- fire_service_calls_tbl_cache is a view
-- cache command is executable only once
-- lazy implies that the table wont be cached until used for the first time


drop view if exists fire_service_calls_tbl_cache;

cache lazy table fire_service_calls_tbl_cache as
select * from demo_db.fire_service_calls_tbl
```

```sql
/*
Caching guidelines:
Cache tables:
1. medium size tables which can fit into spark memory
2. tables that are frequently used

You should not cache:
1. Tiny tables
2. Huge tables that cannot fit inside memory since memory management overhead with overpower caching benefits.
*/
```
