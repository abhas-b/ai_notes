```python
schema = """id int, name string, dop string, phone long, amount string, discount string"""

data_list = [(100, "Prashant", '2020-06-15', 923122311, "12000", 18.5),
             (101, "David", '2018-08-7', 424242424242, "15000", 'nil'),
             (102, 'Simram', '14-05-2025', None, 3535353535, 21)]
```

```python
sales_df = spark.createDataFrame(data_list, schema=schema)

display(sales_df)
```

```python
sales_df.printSchema()
```

```python
sales_df.select("dop", "amount", "discount").describe().show()
```

```python
sales_df.selectExpr("string(id)", "cast(id as string)").show()
```
