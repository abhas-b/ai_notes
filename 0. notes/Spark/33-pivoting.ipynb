```python
import pyspark.sql.functions as F

schema = "customer string, product string, quantity int, unit_price double"

sales_df = spark.read\
                .format("csv")\
                    .schema(schema)\
                        .load("/FileStore/tables/sales")

display(sales_df)
```

```python
sales_df.crosstab("customer", "product")\
    .show()
```

```python
sales_df.groupBy("customer")\
    .pivot("product")\
        .agg(F.expr("sum(quantity * unit_price) as total_value"))\
            .show()
```

```python
sales_df.groupBy("customer")\
    .pivot("product", ["milk", "bread"])\
        .agg(F.expr("sum(quantity * unit_price) as total_value"))\
            .show()
```
