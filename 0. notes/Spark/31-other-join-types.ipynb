```python
stores_list = [("S1", "Down Town"),
               ("S2", "Cape Town"),
               ("S3", "New Delhi")]


product_list = [("01", "Scroll Mouse"),
                ("02", "Optical Mouse")]
```


# Cross Join

```python
stores_df = spark.createDataFrame(stores_list).toDF("store_id", "City")
product_df = spark.createDataFrame(product_list).toDF("product_id", "Product Name")


store_product_df = stores_df.join(product_df)
store_product_df.show()
```


# Semi Join and Anti join

1. Anti join: to find whether info in one table is not available in another table. Example: 

Table 1: cross join of stores and products
Table 2: orders table

Task: find out which store is not selling a particular product



```python
order_list = [("01", "S1", "02", 1, 350),
              ["02", "S1", "01", 5, 1000],
              ["03", "S2", "01", 1, 200]]

orders_df = spark.createDataFrame(order_list).toDF("Order ID", "store_id", "product_id", "qty", "UnitPrice")
orders_df.show()
```

```python
join_expr = (store_product_df.store_id == orders_df.store_id)\
    & (store_product_df.product_id == orders_df.product_id)


store_product_df.join(orders_df, join_expr, "left_anti").orderBy("store_id", "product_id").show()
```

```python
join_expr = (store_product_df.store_id == orders_df.store_id)\
    & (store_product_df.product_id == orders_df.product_id)

#shortcut fot join expr: join_expr = ["store_id", "product_id"]
#This works when: both tables have the same column names, join is on equality operator, and you want AND condition on the join

store_product_df.join(orders_df, join_expr, "left_semi").orderBy("store_id", "product_id").show()
```
