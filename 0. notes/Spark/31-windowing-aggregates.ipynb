```python
import pyspark.sql.functions as F


summary_df = spark.read\
                    .format("parquet")\
                        .load("/FileStore/tables/my_invoices")


summary_df.show(5)
```

```python
from pyspark.sql import Window


invoice_window = Window\
                    .partitionBy("Country")\
                        .orderBy("week")\
                            .rowsBetween(Window.unboundedPreceding, Window.currentRow)


summary_df.withColumn("total invoices", F.sum("NumInvoices").over(invoice_window))
summary_df.show()
```
