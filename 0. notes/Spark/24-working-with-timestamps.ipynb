
Spark date is a DateType Object (int32)
timestamp is a TimestampType object (int96)


```python
import pyspark.sql.functions as F

data_list = [
    (2022,5,18),
    (9999,12,31),
    (-9999,1,1),
    (10000,1,1),
    (-10000,1,1),
    (193,5,25),
    (99,5,25),
    (1000,2,29)
]


spark.createDataFrame(data_list)\
    .toDF("Y", "M", "D")\
        .withColumn("valid_dates", F.make_date("Y","M","D"))\
            .show()
```
