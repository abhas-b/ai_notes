```python
ddl_schema = "ID string, FirstName string, LastName string"

df = spark.read\
    .format("json")\
        .schema(ddl_schema)\
            .load('/FileStore/struct_record.json')


display(df)
```

```python
df.schema
```

```python
%fs head /FileStore/struct_of_struct.json
```

```python
ddl_schema = """ID string, FirstName string, LastName string,
                Address STRUCT<AddressLine1 string, AddressLine2 string, City string, State string, Country string, Pin string>"""

df = spark.read\
    .format("json")\
        .schema(ddl_schema)\
            .load('/FileStore/struct_of_struct.json')


display(df)
```

```python
df.printSchema()
```

```python
df.select("FirstName", "Address.City").show()
```
