
# Programmatic Schema

```python
from pyspark.sql.types import *

flight_schema_struct = StructType([
    StructField("FL_DATE", DateType()),
    StructField("OP_CARRIER", StringType()),
    StructField("OP_CARRIER_FL_NUM", IntegerType()),
    StructField("ORIGIN", StringType()),
    StructField("ORIGIN_CITY_NAME", StringType()),
    StructField("DEST", StringType()),
    StructField("DEST_CITY_NAME",StringType()),
    StructField("CRS_DEP_TIME", StringType()),
    StructField("DEP_TIME", IntegerType()),
    StructField("WHEELS_ON", IntegerType()),
    StructField("TAXI_IN", IntegerType()),
    StructField("CRS_ARR_TIME", IntegerType()),
    StructField("ARR_TIME", IntegerType()),
    StructField("CANCELLED", IntegerType()),
    StructField("DISTANCE", IntegerType())
])
```

```python
flight_time_csv_df = spark.read\
                        .format("csv")\
                            .schema(flight_schema_struct)\
                                .option("dateformat", "M/d/y")\
                                    .option("mode", "FAILFAST")\
                                        .option("header", "true")\
                                            .load("/FileStore/tables/flight_time.csv")
flight_time_csv_df.show(5)
```

```python
flight_time_csv_df.schema.simpleString()
```


Spark supports schema on read. Which means, for parquet and avro files if we manually provide schema then its inbuilt schema definition would be overwritten.

```python
flight_time_json_df = spark.read\
                        .format("json")\
                            .load("/FileStore/tables/flight_time.json")
flight_time_json_df.show(5)
```

```python
flight_time_parquet_df = spark.read\
                            .format("parquet")\
                                .load("/FileStore/tables/flight_time.parquet")
flight_time_parquet_df.show(5)
```

```python
flight_time_parquet_df.schema.simpleString()
```

```python
flight_time_avro_df = spark.read\
                            .format("avro")\
                                .load("/FileStore/tables/flight_time.avro")
flight_time_avro_df.show(5)
```

```python
flight_time_avro_df.schema.simpleString()
```
